```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

# Figuring out Misssing Values

```python
missing = np.nan

series_obj = Series(np.arange(8), index=['row1','row2', missing,'row4','row5','row6','row7', missing])
series_obj
```

# Filling Missing Values

```python
np.random.seed(25)
Df_obj = DataFrame(np.random.rand(36).reshape(6,6),
                   index=['row1','row2','row3','row4','row5','row6'],
                  columns=['col1','col2','col3','col4','col5','col6'])
Df_obj
```

```python
Df_obj.loc[3:5,'col1'] = missing
Df_obj.loc[1:4,'col5'] = missing
Df_obj
```

```python
filled_df =Df_obj.fillna(0)
```

```python
filled_df
```

```python
filled_df =Df_obj.fillna({'col1':0.1, 'col5':1.25})
filled_df
```

```python
filled_df1 =Df_obj.fillna(method='ffill')
filled_df1
```

# Count Missing Values

```python
np.random.seed(24)
Df_obj1 = DataFrame(np.random.rand(36).reshape(6,6))
Df_obj1.loc[3:5,0] = missing
Df_obj1.loc[1:4,5] = missing
Df_obj1
```

```python
Df_obj1.isnull().sum()
```
